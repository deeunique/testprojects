```python
from tkinter import *
from tkinter import ttk
from random import randint,choice

class MyApp:
    def __init__(self,master,*args,**kwargs):
        self.master=master
        self.start=ttk.Button(self.master,text="Start").pack()
        self.canvas=Canvas(self.master,width=300,height=300,bg="powder blue")
        self.canvas.pack()
        
        
    def draw_random(self,n):
        color=["red","blue","black","green","powder blue","white"]
        for i in range(n):
            x1=randint(1,300)
            x2=randint(1,300)
            y1=randint(1,300)
            y2=randint(1,300)
            
            self.canvas.create_rectangle(x1,y1,x2,y2,outline=choice(color),width=2)
            
        
        
        
root=Tk()
app1=MyApp(root)
app1.draw_random(1000)
root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from random import randint,choice
from time import sleep

class MyApp:
    def __init__(self,master,*args,**kwargs):
        self.master=master
        self.start=ttk.Button(self.master,command=self.move_infinite,text="Start").pack()
        self.canvas=Canvas(self.master,width=300,height=300,bg="black")
        
        self.canvas.pack()
        init_pos=[-3,3]
        self.x=choice(init_pos)
        self.y=3
        
        self.canvas.create_oval(50,50,60,60,fill="red",outline="red")
        self.canvas.move(1,randint(1,200),randint(1,200))
        self.canvas.create_rectangle(10,260,100,270,fill="white")
        self.master.bind("<Up>",self.move_up)
        self.master.bind("<Down>",self.move_down)
        self.master.bind("<Left>",self.move_left)
        self.master.bind("<Right>",self.move_right)
        
    def move_infinite(self):
        color=["pink","blue","white","black","green"]
        while True:
           
            try:
                position=self.canvas.coords(1)
                if position[3]>=300:
                    self.y=-3
                    
                elif position[2]>=300:
                    self.x=-3
                elif position[1]<=0:
                    self.y=3
                elif position[0]<=0:
                    self.x=3
                
                    
            
                root.update()
                self.canvas.move(1,self.x,self.y)
                sleep(0.02)
            except:
                pass
    def move_up(self,e):
        self.canvas.move(2,0,-10)
    def move_down(self,e):
        self.canvas.move(2,0,10)
    def move_right(self,e):
        self.canvas.move(2,10,0)
    def move_left(self,e):
        self.canvas.move(2,-10,0)
        
        
root=Tk()
app1=MyApp(root)
root.mainloop()
```
